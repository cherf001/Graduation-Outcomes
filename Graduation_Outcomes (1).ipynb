```python
# Import required libraries
from csv import reader

# The reader method returns a reader object which iterates over lines in the file, so its contents can be written
# and read.
data_rdr = reader(open('GO.csv', 'r'))
header_rdr = reader(open('GO.csv', 'r', encoding = "utf8"))

data_rows = [d for d in data_rdr]
header_rows = [h for h in header_rdr if h[0] in data_rows[0]]

all_short_headers = [h[0] for h in header_rows]
skip_index = []
final_header_rows = []

# Create a for loop that itirates through the list to return the values in a dictionary
for header in data_rows[0]:
# # Invoke if/else as conditional statements.
    if header not in all_short_headers:
        index = data_rows[0].index(header)
        skip_index.append(index)
    else:
        for head in header_rows:
            if head[0] == header:
                final_header_rows.append(head)
                break

del all_short_headers
new_data = []
for row in data_rows[1:]:
    new_row = []
    for i, d in enumerate(row):
        if i not in skip_index:
            new_row.append(d)
        new_data.append(new_row)


zipped_data = []
for drow in new_data:
    zipped_data.append(zip(header_rows, drow))
```

```python
# Create a for loop that itirates through the list and returns the values.
for value1, value2 in zipped_data[0]:
    print('Question: {}\nAnswer: {}'.format(value1[0:2], value2))
```

```python
# loading the required libraries and the data
# Import required libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

# Reading the data
df = pd.read_csv('GO.csv')
print(df.shape)
print(df.info())
```

```python
# In the dataset, there are 430 observations of 22 variables. All of the variables have 430 records. This indicates
# that the data has no missing value.
```

```python
# Identify Outliers
df.describe()
```

```python
# There is a lot of outliers when looking at the variables. Outliers are in % of cohort Total Regents,
# # of grads Total Regents, % of cohort Advanced Regents, and so forth as a result of incorrect data.
# I choose # of grads Total Regents variable as the focus of outliers detection.
# In the # of grads Total Regents variable, the minimun and maximum numbers of grads total regents are 35 and
# 13479 respectively. 
```

```python
# Identifying Outliers with Interquartile Range (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
# The output will print the IQR scores, which can be used to detect outliers.
```

```python
# Determining points that indicates the presence ot outliers where a true value is a positive test for outliers.
print(df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))
```

```python
# Fixing outliers- Techhniques like Quantile-based Flooring and Capping,Trimming, IQR Score,
# Log Transformation, and so forth are used to treat outliers.

# Fixing outliers using the IQR Score technique. Anuthing that does not fall in range (Q1 - 1.5 IQR) and (Q3 + 1.5 IQR)
# is an outlier, and it can be removed.
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_out.shape)
```

```python
# The initial dataset was 430 observations of 22 variales. The IQR Score outliers removal technique returns
# 332 observations of 22 variables. This indicates that a lot of records get deleted if the IQR method is used. 
```
